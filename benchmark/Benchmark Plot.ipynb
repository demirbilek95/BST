```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("benchmark_times.txt",sep ="\t", header=None)
data.columns = ["# of Nodes", "std::map", "Unbalanced Tree", "Balanced Tree"]
```

```python
data
```

```python
fig = plt.figure(figsize=(12,8))
plt.ylabel("Time (microsecond)")
plt.xlabel("Number of Nodes")

ax = plt.subplot(111)

ax.plot(data['# of Nodes'], data['std::map'], label = 'std::map')
ax.plot(data['# of Nodes'], data['Unbalanced Tree'], label = 'Unbalanced Tree')
ax.plot(data['# of Nodes'], data['Balanced Tree'], label = 'Balanced Tree')

plt.title("Benchmark Results")
ax.legend()

plt.savefig("Benchmark_results.png")
```
